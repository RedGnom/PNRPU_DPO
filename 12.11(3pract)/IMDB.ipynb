```python
#Обязателен selenium и edge.
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.options import Options
import time
import requests
from bs4 import BeautifulSoup
import random

# Путь к драйверу Edge
edge_driver_path = r""  # На другом пк сменить на путь к web driver(Использовалась версия 130)

# Настройка службы Edge
service = Service(edge_driver_path)

# Настройка опций Edge
options = Options()
options.use_chromium = True  # Используем Chromium-версию Edge

# Настройка Edge driver
driver = webdriver.Edge(service=service, options=options)
driver.get("https://www.imdb.com/chart/top/")

# Ожидание загрузки страницы
time.sleep(5)

# Получение HTML-кода страницы
page_source = driver.page_source

# Закрытие драйвера
driver.quit()

soup = BeautifulSoup(page_source, 'lxml')
films = soup.findAll('div', class_='ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-a69a4297-2 bqNXEn cli-title with-margin')

print(f"Найдено {len(films)} фильмов")

# Предотвращение 403 forbidden
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
random_films = random.sample(films, 10)

for film in random_films:
    title = film.text.strip()
    link = film.find('a')['href'] #Кусок ссылки с imdb
    full_link = f"https://www.imdb.com{link}"  # Полная ссылка
    description_for_film = requests.get(full_link, headers=headers)
    bs = BeautifulSoup(description_for_film.text, 'lxml')

    parent_element = bs.find('p', {'data-testid': 'plot', 'class': 'sc-3ac15c8d-3 bMUzwm'})

    if parent_element:
        first_span = parent_element.find('span', {'role': 'presentation', 'data-testid': 'plot-xs_to_m'})

        if first_span:
            print(f"Название: {title}")
            text = first_span.text.strip()
            print(text)
            print("_"*40)
            print()
        else:
            print(f"Название: {title}")
            print("Описание не найдено")
            print()
```
